```python
#cleaning value
def value_parser(x):
    val=x[1:-1]
    mk=x[-1]
    try:
        if mk=="M":
            return float(val)*1000.0
        else:
            return float(val)
    except ValueError:
        return np.nan
```

```python
df_new=df['Value'].apply(value_parser)
```

```python
np.ceil(np.mean(df_new))
```

```python
df_new=df_new.fillna(np.ceil(np.mean(df_new)))
```

```python
df_new=pd.DataFrame(df_new,columns=['Value'])
```

```python
df_new['Wage'] = df['Wage'].apply(value_parser)
df_new['Wage'].fillna(np.ceil(np.mean(df_new['Wage'])), inplace=True)
```

```python
df['Joined'][df['Joined'].isna()].shape[0]
```

```python
#converting joined "datetime" object
df_new['Joined']=pd.to_datetime(df['Joined'])
#getting current date as pd.series
this_year = pd.to_datetime(pd.Series(['Jan 1, 2019']*df_new['Joined'].shape[0]))
#calculating palyer experience
df_new['experience']=this_year-df_new['Joined']
#dropping joined
df_new.drop('Joined',axis=1,inplace=True)
#converting to string
df_new['experience']=df_new['experience'].astype(str)
```

```python
def exp_parser(x):
    return x.split(' ')[0]
```

```python
df_new['experience'] = df_new['experience'].apply(exp_parser)
```

```python
def exp_to_num(x):
    try:
        return int(x)
    except ValueError:
        return np.nan
```

```python
df_new['experience']=df_new['experience'].apply(exp_to_num)
df_new['experience'].fillna(np.ceil(np.mean(df_new['experience'])),inplace=True)
```

```python
#cleaning weight
df['Weight'].unique()
```

```python
def wght(x):
    try:
        return int(x[:-3])
    except TypeError:
        return np.nan
```

```python
df_new['Weight']=df['Weight'].apply(wght)
```

```python
print("No of Nan Values in weight feature is:",df_new['Weight'][df_new['Weight'].isna()].shape[0])
```

```python
df_new['Weight'].fillna(np.ceil(np.mean(df_new['Weight'])),inplace=True)
```

```python
#cleaning height
df['Height'].unique()
```

```python
def parse_height_to_cm(x):
    try:
        ft,inch = x.split("'")
        return (int(ft)*12+int(inch))*2.54
    except:
        np.nan
    
```

```python
df_new['Height']=df['Height'].apply(parse_height_to_cm)
print('Number of Nan values in `Height` is:', df_new['Height'][df_new['Height'].isna()].shape[0])
```

```python
df_new['Height'].fillna(np.ceil(np.mean(df_new['Height'])),inplace=True)
```
